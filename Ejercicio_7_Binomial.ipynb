```python
from scipy.stats import binom
import matplotlib.pyplot as plt
```

### Ejercicio 7. 
*Seguridad.* Según un anuncio comercial de colchones, 30% de los accidentes automovilísticos se deben a la falta de sueño. Supón que en un día cualquiera ocurren 10 accidentes automovilísticos.  

**a) Determina la distribución de probabilidad y traza la gráfica correspondiente.**  
Dado que estamos analizando un número fijo de accidentes (10) y cada uno tiene una probabilidad independiente (0.3) de ser causado por la falta de sueño, estamos tratando con una distribución binomial.  
En este caso cada ensayo de Bernoulli tiene dos resultados posibles (accidente por falta de sueño o no), y la probabilidad de éxito (accidente por falta de sueño) es constante (0.3).

```python
################## Inciso a) ##################
n, p = 10, 0.3
k = list(range(n + 1))
y = binom.pmf(k, n, p)

plt.bar(k, y, color=['#12229D'])
plt.xlabel("Número de accidentes por falta de sueño")
plt.ylabel("Probabilidades")
plt.title("Distribución Binomial")
plt.show()
```

**b)** ¿Qué tipo de sesgo presenta la distribución?  
La gráfica de una distribución binomial, por ser una función discreta, se compone de columnas, donde la altura representa la probabilidad asociada a cada valor de $X$.   
Para este caso la gráfica presenta un sesgo, según el valor de $p$.  
Para la gráfica anterior $p < 0.5$, la gráfica está sesgada a la izquierda.  
Donde cada accidente tiene una probabilidad independiente ($p$) igual a $0.3$.

**c) Calcula la probabilidad de que cinco de los accidentes se hayan debido a la falta de sueño.**  
Debido a que nos pide la probabilidad de que exactamente 5 de los accidentes hayan debido a la falta de sueño tomamos como casos de exito $k = 5$ en la distribucion de probabilidad binomial.
Función de probabilidad: 
$$
\begin{align}
    P(k; n, p) = \binom{n}{k} p^k (1 - p)^{n - k}
\end{align}
$$  
Calculando la probabilidad de 5 accidentes:
$$
\begin{align}
    P(5; 10, 0.3) &= \binom{10}{5} {0.3}^5 (1 - {0.3})^{10 - 5} \\\\
    P(5; 10, 0.3) &= 0.1029
\end{align}
$$  
Por lo tanto, la probabilidad de que cinco de los accidentes se hayan debido a la falta de sueño es: **0.1029**

```python
################## Inciso c) ##################
n, p, k = 10, 0.3, 5
y = binom.pmf(k, n, p)
print('c)',y)
```

**d)** Calcula la probabilidad de que al menos 2 accidentes hayan sido ocasionados por la falta de sueño.  
La probabilidad de que al menos dos de los accidentes hayan debido a la falta de sueño tomamos como casos de exito $k = {2,3,4,5,6,7,8,9,10}$ en la distribucion de probabilidad binomial $P(k ≥ 2)$.  
Función de probabilidad: 
$$
\begin{align}
    P(k; n, p) = \binom{n}{k} p^k (1 - p)^{n - k}
\end{align}
$$  
Calculando la probabilidad de al menos 2 accidentes accidentes:
$$
\begin{align}
    P((k ≥ 2); 10, 0.3) &= \sum_{k=2}^{10} \binom{10}{k} (0.3)^k (0.7)^{10-k} \\\\
    P((k ≥ 2); 10, 0.3) &= 0.8506
\end{align}
$$  
Por lo tanto, la probabilidad de que al menos 2 accidentes de los accidentes se hayan debido a la falta de sueño es: **0.8506**

```python
################## Inciso d) ##################
n, p = 10, 0.3
k = range(2, n+1)
y = binom.pmf(k, n, p)
print('d)',sum(y))
```

**e) ¿Cuántos accidentes esperarías que fueran provocados por falta de sueño?**  
La media o valor esperado de para una distribución de probabilida binomial se calcula de la siguiente forma:
$$
\begin{align}
    E[X] = n \cdot p
\end{align}
$$  
De esta forma:
$$
\begin{align}
    E[X] = 10 \cdot 0.3 = 3
\end{align}
$$  
Por lo tanto, se esperarían que **3** de los accidentes fueran provocados por falta de sueño.

```python
################## Inciso d) ##################
n, p = 10, 0.3
EX = n * p
print('e)',EX)
```
